### 2차 함수 그래프 with plot()

```python
# 리스트를 이용해서 1차 함수 y = x를 그려보면 :

plt.plot([0, 1, 2, 3, 4])
plt.show()
```

```python
# numpy.array를 이용해서 함수 그래프 그리기
# y = x^2

x = np.array([1, 2, 3, 4, 5]) # x = 정의역
y = np.array([1, 4, 9, 16, 25]) # y = f(x)

plt.plot(x, y)
plt.show()
```

위의 그래프를 보다 더 부드럽게 만들기 위해 무수히 많은 점을 찍어보자. 

```python
# np.arrange(a, b, c) c : 0.01

x = np.arange(-10, 10, 0.01)
plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축에 설명 추가하기 

x = np.arange(-10, 10, 0.01)

###
plt.xlabel("x value")
plt.ylabel("f(x) value")
###

plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축의 범위를 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5, 5, 0, 25]) #[x-min, x_max, y_min, Y_max]

plt.plot(x, x**2)
plt.show()
```

```python
# x, y 축에 눈금 설정하기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) #[x-min, x_max, y_min, Y_max]

# tick이 눈금 하나를 의미
plt.xticks([i for i in range(-5, 5, 1)]) # x축의 눈금 설정, -5, -4, -3 , ...
plt.yticks([i for i in range(0, 27, 3)]) # y 축의 눈금 설정, 

plt.plot(x, x**2)
plt.show()
```

```python
# 그래프에 title 달기, 범례 달기

x = np.arange(-10, 10, 0.01)

plt.xlabel("x value")
plt.ylabel("f(x) value")
plt.axis([-5, 5, 0, 25]) 
plt.xticks([i for i in range(-5, 5, 1)]) 
plt.yticks([i for i in range(0, 27, 3)]) 

###
plt.title("y = x^2 graph")
###

plt.plot(x, x**2, label = "trend") # 그래프의 선을 trend라고 지정

### legned를 그리는 시점은 반드시 선을 그린이후가 되야함.
plt.legend()
###

plt.show()
```

## 3. Matplotlib Case Study 

### 꺾은선 그래프(Plot)

- 시계열 그래프에서 많이 사용

```python
x = np.arange(20) # 0 ~ 19
y = np.random.randint(0, 20, 20) #난수를 20번 생성

x, y 
```

```python
plt.plot(x, y)

# y축을 20까지 보이게 하고싶다면?
plt.axis([0, 20, 0, 20])

# y축을 5단위로 보이게 하고 싶다면?
plt.yticks([0, 5, 10, 15, 20])

plt.show()

```

### 산점도 (Scatter Plot)

- x와 y가 별개의 그래프 일 때 사용
- '.scatter()'

```python
plt.scatter(x, y)
plt.show()
```

### 박스 그림 (Box Plot)

- 수치형 데이터에 대한 정보(Q1, Q2, Q3, min, max)
- Q1 : 25%에 해당하는 부분 
- Q2 : 상위 50%에 해당하는 부분, 중앙값
- Q3 : 75%에 해당하는 부분
- 데이터의 *축약*을 나타냄

```python
plt.boxplot((x, y)) #
plt.title("Box plot of x, y")

plt.show()
```

### 막대 그래프 (Bar Plot)

- 범주형 데이터의 "값"과 그 크기를 직사각형으로 나타낸 그림
- *도수*를 표현하기 용이

```python
plt.bar(x, y)

# xticks를 올바르게 처리하기.
plt.xticks(np.arange(0,20,1))

plt.show()
```

### cf) Histogram

- 도수분포를 직사각형의 막대 형태로 나타냈다.
- **계급**으로 나타낸 것이 특징 
- **범주형** 데이터로 구성 후 그림을 그림
- 막대를 붙여서 나타냄
- plt.hist()

```python
plt.hist(y, bins = np.arange(0,20,2))

# xticks를 올바르게 고치기
plt.xticks(np.arange(0,20,2))

plt.show()
```

### 원형 그래프 (Pie Chart)

- 데이터에서 전체에 대한 부분의 비율을 부채꼴로 나타낸 그래프
- 다른 그래프에 비해서 *비율* 확인에 용이
- '.pie()'

```python
z = [100, 300, 200, 400]

plt.pie(z, labels = ['one', 'two', 'three', 'four']) # 데이터와 대응되는 label 붙이기
plt.show()
```

## 4. Seaborn

### Matplotlib를 기반으로 더 다양한 시각화 방법을 제공하는 라이브러리
- 커널밀도 그림

### Seaborn Import하기

```python
import seaborn as sns
```

### 커널밀도그림 (Kernel Density Plot)

- 히스토그램과 같은 연속적인 분포를 곡선화해서 그린 그림
- sns.kdeplot()

```python
# in Histogram

x = np.arange(0, 22, 2)
y = np.random.randint(0, 20, 20)

plt.hist(y, bins = x)
plt.show()
```

```python
# kdeplot

sns.kdeplot(y, shade = True) # 히스토그램의 분포가 스무딩하게 표현됨. / shade : 선 밑에 음영을 줌

plt.show()
```

### 카운트그림 (Count Plot)

- 범주형 column의 빈도수를 시각화 -> Groupby 후의 도수를 하는 것과 동일한 효과
- sns.countplot()

```python
vote_df = pd.DataFrame({"name" : ['Andy', 'Bob', "Cat"], "vote" : [True, True, False]})

vote_df
```

```python
# in matplotlib barplot

vote_count = vote_df.groupby('vote').count()

vote_count
```

```python
plt.bar(x = [False, True], height = vote_count['name'])

plt.show()
```

```python
# sns의 countplot

sns.countplot(x = vote_df['vote']) # x 축을 범주형으로 바꾸어줌, countplot : 색을 다르게 해줌

plt.show()
```

### 캣그림(Cat Plot)

- 숫자형 변수와 하나 이상의 범주형 변수의 관계를 보여주는 함수
- sns.catplot()
- 여러개의 데이터를 묶는 하나의 방법

```python
covid = pd.read_csv("C:/Users/kmusw/.jupyter/country_wise_latest.csv")

covid.head(5)
```

```python
s = sns.catplot(x ="WHO Region", y ='Confirmed' , data = covid, kind = 'violin')  # violin, strip
s.fig.set_size_inches(10,6)
plt.show()
```

### 스트립그림 (Strio Plot)

- scatter plot과 유사하게 데이터의 수치를 표현하는 그래프
- sns.stripplot()

```python
sns.stripplot(x= "WHO Region", y="Recovered", data= covid)

plt.show()
```

### cf) swarmplot

- 유사한 점들이 겹치는 경우 양옆으로 분산해줌
- 값들을 파악하기 더 쉬움

```python
sns.swarmplot(x= "WHO Region", y="Recovered", data= covid)

plt.show()
```

### 히트맵 (Heatmap)

- 데이터의 행렬을 색상으로 표현해주는 그래프
- sns.heatmap()

```python
# 히트맵 예제

covid.corr()
```

```python
sns.heatmap(covid.corr())

plt.show()
```
