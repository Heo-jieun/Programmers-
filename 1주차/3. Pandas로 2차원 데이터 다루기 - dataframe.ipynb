### Pandas활용 4. 행을 기준으로 데이터 접근하기

```python
# 예시 데이터 - 도서과 정보

books_dict = {"Available" : [True, True, False], 'Location' : [102, 215, 323], "Genre" : ["Programming", "Physics", "Math"]}

books_df = pd.DataFrame(books_dict, index = ['버그란 무엇인가', '두근두근 물리학', '미분해줘 홈즈'] ) # 인덱스 이름을 지정

books_df
```

### 인덱스를 이용해서 가졍기 : '.loc[row, col]'

```python
books_df.loc['버그란 무엇인가'] # -> series type
```

```python
# "미분해줘 홈즈 책이 대출 가능한지?"

books_df.loc['미분해줘 홈즈']['Available'] # 첫번쨰 방법

books_df.loc['미분해줘 홈즈', 'Available'] # 두번째 방법
```

### 숫자 인덱스를 이용해서 가져오기 : '.iloc[rowidx, colidx]'

```python
# 인덱스 0행의 인덱스 1열 가져오기

books_df.iloc[0,1] # numpy 의 array에서 사용했던 방식과 완전 동일
```

```python
# 인덱스 1행의 인덱스 0~1열 가져오기

books_df.iloc[1, 0:2]
```

## Pandas 활용 5. groupby

- Split : 특정한 '기준'을 바탕으로 DataFrame을 분활
- Apply : 통계함수 - sum(), mean(), median(), 등을 적용해서 각 데이터를 압축
- Combine : Apply된 결과를 바탕으로 새로운 Series를 생성 (group_key :applied_value)

.groupby()

```python
covid.head(5)
```

```python
# WHO Region 별 확진자수 

# 1. covid에서 확진자 수 colum만 추출한다.
# 2. 이를 covid의 WHO Region을 기준으로 groupby한다. 

covid_by_region = covid['Confirmed'].groupby(by = covid["WHO Region"])

covid_by_region
```

```python
covid_by_region.sum() # 각 region별로 합쳐줌.
```

```python
# 국가별 감염자 수 

covid_by_region.mean() # sum() / 국가 수 
```
